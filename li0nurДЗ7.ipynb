```python
plt.figure(figsize=(10, 6))
plt.xlabel(u'ось Х', family='verdana', fontweight = 'bold')
plt.ylabel(u'ось У', family='verdana', fontweight = 'bold', rotation='horizontal')
plt.plot(range(-10,11), y1, label='$y=x^3$')
plt.plot(range(-10,11), y2, label='$y=2 * x^3$')
plt.fill_between(range(-10,11), y1, y2, color='green', alpha=0.5, label=u'заливка')
plt.legend(loc=2)
```

Для этого нужно изобразить объекты этих классов разными цветами в координатах соответсвующих пар признаков (для работы с цевтом подойдет ListedColormap)  и проанализировать полученные изображения. 

На одном полотне отобразите объекты в координатах 'mean radius' и 'mean texture', а на другом - в координатах 'mean perimeter' и 'mean area'.
